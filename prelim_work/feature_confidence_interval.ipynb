```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('../data/Boats_Cleaned_dataset.csv')
df = df.dropna()
df.shape
```

```python
df = df.drop(columns=['Unnamed: 0', 'id', 'created_date', 'sellerId', 'zip', 'city', 'state', 'created_month', 'created_year', 'engineCategory', 'hullMaterial', 'fuelType', 'model', 'make'])
df.head()
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
stnd = StandardScaler().set_output(transform='pandas')
df = stnd.fit_transform(df)
```

```python
n_samples = 1000

lr = LinearRegression()
cols = df.drop('price', axis=1).columns
coef = pd.DataFrame(index=cols, columns=range(n_samples))
for k in range(n_samples):
    df_b = df.sample(frac=1.0, replace=True)
    X = df_b.drop('price', axis=1)
    y = df_b.price
    lr.fit(X,y)
    coef.iloc[:,k] = lr.coef_
```

```python
quantiles = coef.quantile([0.05,0.95],axis=1).T.round(3)
quantiles
```
